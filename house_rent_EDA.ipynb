# Outlier analysis

```python
data = rent_df
```

```python
fig, axs = plt.subplots(2,3, figsize = (10,5))
plt1 = sns.boxplot(data['Rent'], ax = axs[0,0])
plt2 = sns.boxplot(data['Size'], ax = axs[0,1])
plt.tight_layout()
```

```python
# outlier treatment for price

plt.boxplot(data.Rent)
Q1 = data.Rent.quantile(0.25)
Q3 = data.Rent.quantile(0.75)
IQR = Q3 -Q1
housing = data[(data.Rent >= Q1 -1.5*IQR) & (data.Rent <= Q3 + 1.5*IQR)]
housing.shape
```

```python
# outlier detection

max_threshold = data['Rent'].quantile(0.9)
min_threshold = data['Rent'].quantile(0.005)

df = data[ (data.Rent > min_threshold) & (data.Rent < max_threshold) ]
df.shape
```

```python
plt.boxplot(data.BHK)
Q1 = data.BHK.quantile(0.25)
Q3 = data.BHK.quantile(0.75)
IQR = Q3 -Q1
housing = data[(data.BHK >= Q1 -1.5*IQR) & (data.BHK <= Q3 + 1.5*IQR)]
```

```python
plt.boxplot(data.Size)
Q1 = data.Size.quantile(0.25)
Q3 = data.Size.quantile(0.75)
IQR = Q3 -Q1
housing = data[(data.Size >= Q1 -1.5*IQR) & (data.Size <= Q3 + 1.5*IQR)]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import category_encoders as ce
```

```python
#Assuming X is your feature set and y is your labels
X = df.drop('Rent', axis = 1)
y = df.Rent
```

```python
encoder = ce.LeaveOneOutEncoder()
X = encoder.fit_transform(X,y)
```

```python
# Splitting the dataset into 80% training and 20% testing
X_train,X_test,y_train,y_test = train_test_split(X,y, train_size = 0.2, random_state = 10)
```

```python
model_rfr = RandomForestRegressor(n_estimators = 50)
model_dtr = DecisionTreeRegressor()
model_lr = LinearRegression()
```

```python
models =[model_rfr, model_dtr, model_lr]
```

```python
for model in models:
    print("fittinf model : %s"%(model))
    model.fit(X_train, y_train)
```

```python
for model in models:
    print(f'Score of {model} of training data   : {model.score(X_train, y_train)}')  
```

```python
for model in models:
    print(f'Score of {model} of testing data   : {model.score(X_test, y_test)}')  
```

```python
#predict 
for model in models:
    y_pred = model.predict(X_test)
    print(f'Report: {model}')
```

```python
def regression_results(y_true, y_pred):
    #regression metrics
    #explained_varience = metrics.explained_varience(y_true,y_pred)
    mean_absolute_error = metrics.mean_absolute_error(y_true,y_pred)

    mse = metrics.mean_squared_error(y_true,y_pred)
    #mean_squared_log_error = metrics.mean_squared_log_error(y_true,y_pred)
    #median_absolute_error = metrics.median_absolute_error(y_true,y_pred)
    r2 = metrics.r2_score(y_true,y_pred)

    #print("explained_varience :", round(explained_varience,4))
    #print('mean_squared_log_error: ', round(mean_squared_log_error,4))
    print('r2: ',round(r2,4))
    print('MAE: ', round(mean_absolute_error,4))
    print('MSE: ', round(mse,4))
    print('RMSE: ', round(np.sqrt(mse),4))
    #print('median_absolute_error', round(median_absolute_error,4))
    
```

```python
for model in models:
    y_pred = model.predict(X_test)
    
    print(f'Report: {model}')
    print(f'{regression_results(y_test, y_pred)}\n')
```
